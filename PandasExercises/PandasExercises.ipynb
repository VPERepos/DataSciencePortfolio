# 23. Solve following:
In the cell below, you have a DataFrame df that consists of 10 columns of floating-point numbers. Exactly 5 entries in each row are NaN values.

For each row of the DataFrame, find the column which contains the third NaN value.

You should return a Series of column labels: e, c, d, h, d

```python
nan = np.nan

data = [[0.04,  nan,  nan, 0.25,  nan, 0.43, 0.71, 0.51,  nan,  nan],
        [ nan,  nan,  nan, 0.04, 0.76,  nan,  nan, 0.67, 0.76, 0.16],
        [ nan,  nan, 0.5 ,  nan, 0.31, 0.4 ,  nan,  nan, 0.24, 0.01],
        [0.49,  nan,  nan, 0.62, 0.73, 0.26, 0.85,  nan,  nan,  nan],
        [ nan,  nan, 0.41,  nan, 0.05,  nan, 0.61,  nan, 0.48, 0.68]]

columns = list('abcdefghij')

df = pd.DataFrame(data, columns=columns)
```

The answer should be a Series as follows:<br/>

A<br/>

# 26. Solve following:
Consider a DataFrame df where there is an integer column 'X':

df = pd.DataFrame({'X': [7, 2, 0, 3, 4, 2, 5, 0, 3, 4]})

For each value, count the difference back to the previous zero (or the start of the Series, whichever is closer). These values should therefore be

[1, 2, 0, 1, 2, 3, 4, 0, 1, 2]

Make this a new column 'Y'.

```python
df = pd.DataFrame({'X': [7, 2, 0, 3, 4, 2, 5, 0, 3, 4]})

new_row = pd.DataFrame([[0]], columns=df.columns)

df = pd.concat([new_row, df], ignore_index=True)

groups = df['X'].eq(0).cumsum()

df['Y'] = df.groupby(groups).cumcount()
df = df.drop(0)
df = df.reset_index(drop=True)

print(df['Y'])
```

# 28. Solve following:
You are given the DataFrame below with a column of group IDs, 'grps', and a column of corresponding integer values, 'vals'. <br/>

df = pd.DataFrame({"vals": np.random.RandomState(31).randint(-30, 30, size=15), <br/>
                   "grps": np.random.RandomState(31).choice(["A", "B"], 15)})<br/>

Create a new column 'patched_values' which contains the same values as the 'vals' any negative values in 'vals' with the group mean.

```python
df = pd.DataFrame({"vals": np.random.RandomState(31).randint(-30, 30, size=15),
"grps": np.random.RandomState(31).choice(["A", "B"], 15)})
print(df)
```

```python
group_means = df[df['vals'] > 0].groupby('grps')['vals'].mean()

df['patched_values'] = df['vals'].where(
    df['vals'] > 0,
    df['grps'].map(group_means)
)

print(df)
```
