# 1. Initialize numpy array and a python list with 1 million integers and multiply each element by a constant.

```python
import numpy as np

numpy_array = np.arange(int(1e6))

a_list = list(range(int(1e6)))

%timeit -n 100 numpy_array2 = numpy_array*2
```

```python
%timeit -n 10 a_list2 = [x * 2 for x in a_list]
```
