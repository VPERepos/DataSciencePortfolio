# 53. Considering a one-dimensional vector D, how to compute means of subsets of D using a vector S of same size describing subset indices?

````python
array = np.array([10, 20, 30, 40, 50])
group_indeces = np.array([0, 1, 1, 0, 2])  # group indices

# sum per group
sums = np.bincount(group_indeces, weights=array)

# count per group
counts = np.bincount(group_indeces)

# mean per group
means = sums / counts

print(means)
````

# 55. Consider the vector [1, 2, 3, 4, 5], how to build a new vector with 3 consecutive zeros interleaved between each value?

````python
array = np.array([1,2,3,4,5])
zero_array = np.array([0,0,0,0,0])

array = array.reshape((-1,1))
zero_array = zero_array.reshape((-1,1))
zero_arrays = np.tile(zero_array, 3)
array = np.concatenate((array, zero_arrays), axis = 1)
array = array.ravel()

print(array)
````

# 58. Consider a set of 10 triplets describing 10 triangles (with shared vertices), find the set of unique line segments composing all the triangles.

````python
rng = np.random.default_rng()

vertices = rng.random((8, 2))
triangles = rng.integers(0, 8, size=(10, 3))
tri_coords = vertices[triangles]

lines = np.concatenate([
    np.concatenate([tri_coords[:, 0, :], tri_coords[:, 1, :]], axis=1),
    np.concatenate([tri_coords[:, 1, :], tri_coords[:, 2, :]], axis=1),
    np.concatenate([tri_coords[:, 2, :], tri_coords[:, 0, :]], axis=1),
], axis=0)

print(tri_coords)
print()
print("All lines number:" + str(lines.shape[0]))

edges = lines.reshape(-1, 2, 2)

edges_sorted = np.sort(edges, axis=1)

edges_flat = edges_sorted.reshape(-1, 4)

unique_lines = np.unique(edges_flat, axis=0)

print("Unique lines number: " + str(unique_lines.shape[0]))
````
