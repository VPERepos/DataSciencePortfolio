# 17. Create a 8x8 matrix and fill it with a checkerboard pattern.

```python
matrix5 = np.zeros((8,8))
array_checker_1d = np.ones(8)
array_checker_1d[::2] = 0
matrix5[:] = array_checker_1d
matrix5[::2] = array_checker_1d[::-1]
print(matrix5)
```

# 18. Consider a (6,7,8) shape array, what is the index (x,y,z) of the 100th element?

NumPy arrays are stored in row-major order (C-order), meaning the last index changes fastest,so <br/>
* x dimension stride $= 7 \times 8 = 56$
* y dimension stride $= 8$
* z dimension stride $= 1$
So the answer is: <br/>
$(1,5,3)$

# 19. Create a checkerboard 8x8 matrix using the tile function.

```python
pattern_array = np.array([[1,0],[0,1]])
matrix_checker = np.zeros((8,8))
matrix_checker = np.tile(pattern_array, (4,4))
print(matrix_checker)
```
